```python


# Set starting page
base_url = 'https://www.gadgets360.com/laptops/price-list-india?facet^[brand^]=Dell,HP,Acer,Asus,Lenovo,Apple&sort=release_priority,product_id&order=desc&is_ajax=1&v=1&size=100&page={}'
page = 1
laps=[]
while True:
    # Update the page number in the URL
    url = base_url.format(page)
    
    # Send the request
    response = requests.get(url, cookies=cookies, headers=headers)
    
    # Check if the response is successful
    if response.status_code != 200:
        print(f"Failed to fetch page {page}")
        break
    
    # Parse the JSON response (if it's in JSON format)
    data = response.json()
    contents=data['response']
    soup = BeautifulSoup(contents, 'html.parser')
    laptops = soup.find_all('div',class_='_lpdwgt _flx pdbtlinks')
    if len(laptops) < 1:
        break
    # If the data is empty or there are no laptops in the response, stop
    # If less than 100 laptops are returned, it's likely the last page

    # Create an empty list to store laptop details
    laptop_data = []
    for laptop in laptops:
        # Extract name
        name_tag = laptop.find('h3', class_='_hd')
        name = name_tag.text.strip() if name_tag else 'N/A'

        # Extract link
        link_tag = laptop.find('a', title=name)
        link = link_tag['href'] if link_tag else 'N/A'

        # Extract price
        price_tag = laptop.find('div', class_='_lprc')
        price = price_tag.text.strip() if price_tag else 'N/A'

        # Extract specifications
        specs = {}

        # Extract Display Size
        display_size_tag = laptop.find('td', class_='_ttl', text='Display size')
        specs['Display Size'] = display_size_tag.find_next('td').text.strip() if display_size_tag else 'N/A'

        # Extract Display Resolution
        display_res_tag = laptop.find('td', class_='_ttl', text='Display resolution')
        specs['Display Resolution'] = display_res_tag.find_next('td').text.strip() if display_res_tag else 'N/A'

        # Extract Processor
        processor_tag = laptop.find('td', class_='_ttl', text='Processor')
        specs['Processor'] = processor_tag.find_next('td').text.strip() if processor_tag else 'N/A'

        # Extract OS
        os_tag = laptop.find('td', class_='_ttl', text='OS')
        specs['OS'] = os_tag.find_next('td').text.strip() if os_tag else 'N/A'

        # Extract SSD
        ssd_tag = laptop.find('td', class_='_ttl', text='SSD')
        specs['SSD'] = ssd_tag.find_next('td').text.strip() if ssd_tag else 'N/A'

        # Extract Weight
        weight_tag = laptop.find('td', class_='_ttl', text='Weight')
        specs['Weight'] = weight_tag.find_next('td').text.strip() if weight_tag else 'N/A'

        # Create a dictionary for the laptop
        laptop_info = {
            'Name': name,
            'Link': link,
            'Price': price,
            **specs
        }

        # Append the laptop info to the list
        laptop_data.append(laptop_info)
    laps.extend(laptop_data)
    # Increment the page number for the next iteration
    page += 1
```

```python
import requests
from bs4 import BeautifulSoup

# Assuming 'laps' contains the list of laptop data you've already collected

for laptop in laps:
    # Send a request to the laptop's detail page
    if len(laptop['Link'])<5:
        detailed_specs = {}
        laptop['Detailed Specs'] = detailed_specs
        continue
    response = requests.get(laptop['Link'], cookies=cookies, headers=headers)
    
    # Check if the response is successful
    if response.status_code != 200:
        print(f"Failed to fetch details for {laptop['Name']}")
        continue
    
    # Parse the page
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the specifications from the specs section
    specs_section = soup.find('div', {'id': 'specs'})
    
    if specs_section:
        # Initialize a dictionary to store detailed specs
        detailed_specs = {}
        
        # Find all specification tables in the specs section
        tables = specs_section.find_all('div', class_='_gry-bg _spctbl _ovfhide')
        
        for table in tables:
            # Find the category header
            category = table.find('div', class_='_hd').text.strip()
            
            # Extract key-value pairs from each table
            rows = table.find_all('tr')
            for row in rows:
                cells = row.find_all('td')
                if len(cells) == 2:
                    key = cells[0].text.strip()
                    value = cells[1].text.strip()
                    detailed_specs[key] = value
        
        # Add detailed specs to the laptop information
        laptop['Detailed Specs'] = detailed_specs
    else:
        print(f"No specs section found for {laptop['Name']}")

# Now `laps` contains all laptop details along with their detailed specifications
```
